```python
# from tensorflow.keras.models import load_model
# import pickle
# import numpy as np
# model = load_model(r"C:\Users\venug\Downloads\55\38\VideoClassificationModel.keras")
# with open(r"C:\Users\venug\Downloads\55\vc\videoclassificationbinarizer.pickle", "rb") as f:
#     lb = pickle.load(f)
# outputvideo = r"C:\Users\venug\Downloads\55\vc\demo.mp4"
# # Assuming you have a preprocessing step that requires a mean array
# mean = np.array([123.68, 116.779, 103.939], dtype="float32")
# Queue=deque(maxlen=128)

import cv2
import numpy as np
from tensorflow.keras.models import load_model
import pickle
from collections import deque

def process_video(video_path):
    # Load your model and label binarizer here
    model_path = r"C:\Users\venug\Downloads\55\38\VideoClassificationModel.keras"
    label_binarizer_path = r"C:\Users\venug\Downloads\55\vc\videoclassificationbinarizer.pickle"

    model = load_model(model_path)
    with open(label_binarizer_path, 'rb') as f:
        lb = pickle.load(f)

    # Initialize variables
    mean = np.array([123.68, 116.779, 103.939], dtype="float32")
    Queue = deque(maxlen=128)

    # Initialize the video capture object
    capture_video = cv2.VideoCapture(video_path)

    # Check if the video was opened successfully
    if not capture_video.isOpened():
        print("Error opening video file")
        return "Error opening video file"

    while True:
        (taken, frame) = capture_video.read()
        if not taken:
            break

        # Preprocess the frame
        frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
        frame = cv2.resize(frame, (224, 224)).astype("float32")
        frame -= mean

        # Make predictions with the model
        preds = model.predict(np.expand_dims(frame, axis=0))[0]
        Queue.append(preds)

        # Aggregate predictions (e.g., by averaging)
        results = np.array(Queue).mean(axis=0)
        i = np.argmax(results)
        label = lb.classes_[i]

        # Optionally, you can return the label immediately or after processing all frames
        # For simplicity, we'll return the label after processing the first frame
        return label

    # Release the video capture object
    capture_video.release()

    # If the video is empty or no frames were processed, return an error message
    return "No frames processed"

```
